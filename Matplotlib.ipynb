```python
plt.bar(np.arange(sesson.shape[0]) - 0.2 , sesson['2015'], width = 0.2 , color = "Red")
plt.bar(np.arange(sesson.shape[0]), sesson['2016'], width = 0.2, color = "Orange")
plt.bar(np.arange(sesson.shape[0]) + 0.2 , sesson['2017'], width = 0.2 , color = "Yellow")

plt.xticks(np.arange(sesson.shape[0]), sesson['batsman'])

plt.show
```

```python
# stcked bar chart
```

```python
plt.bar(sesson['batsman'], sesson['2015'], color = 'Red', label = '2015')
plt.bar(sesson['batsman'], sesson['2016'] , bottom = sesson['2015'], color = ['Yellow'], label = '2016')
plt.bar(sesson['batsman'], sesson['2017'], bottom = (sesson['2016']+sesson["2015"]), color = 'Orange', label = '2017')

plt.legend()
plt.show
```
